# Project - Pakistan Violent & Threatening Healthcare

In this project, we will be looking at two datasets that include information on violent and threatening incidents that impact aid operations, civilians, education, healthcare, refugees and IDPs in Pakistan. All of this is to make sure that the staff is safe and that hopefully there will be better response outcomes in the future. The link to have access to the two datasets is [here](https://www.kaggle.com/datasets/amaanfaheem/pakistan-violent-and-threatening-incidents#)

Some of the goals that we want to work with for the 2016-2022 dataset are the following:
1. **Which years have the most crime? Of the years that have the most crime, which year was worst?**

For the 2019 dataset, we can focus on these goals:
1. **Which months were there the most crime? Of the months that had the most crime, which month was worse?**
